```python
import os, glob, random, shutil
```

```python
pos = glob.glob("datasets/train/Parasitized/*.png")
neg = glob.glob("datasets/train/Uninfected/*.png")

random.shuffle(pos)
random.shuffle(neg)

val_len = int(len(pos) * .2)
```

```python
val_pos = pos[:val_len]
val_neg = neg[:val_len]

print(len(val_pos))
print(len(val_neg))
```

```python
for i in range(len(val_pos)):
    print(val_pos[i])
    shutil.move(val_pos[i], val_pos[i].replace("/train", "/val"))
    
    print(val_neg[i])
    shutil.move(val_neg[i], val_neg[i].replace("/train", "/val"))
```
