```python
# import json
import numpy as np
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from skimage.filters import threshold_otsu
from skimage.io import imread, imsave
from skimage.color import rgb2gray
from skimage.measure import label, regionprops
import json
import os
import uuid
from PIL import Image
import random
import math
import imageio
import imgaug as ia
from imgaug import augmenters as iaa
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
import glob
```

```python
# Training and Validate Sets
degs = [45,90,135,180,225]
count = 1
species =  {
    "train": {
        "negative": len(glob.glob("../datasets/train/negative/negative-*.jpg"))+1,
        "pf": 1,
        "pm": 1,
        "po": 1,
        "pv": 1
    },
    "val": {
        "negative": len(glob.glob("../datasets/val/negative/negative-*.jpg"))+1,
        "pf": 1,
        "pm": 1,
        "po": 1,
        "pv": 1
    }
}
specie = "pv"

phase = "train"

labels = os.path.join('training.json')
with open(labels,'r') as json_file:
    datas = json.load(json_file)
    for data in datas:
        read_path = (data['image']['pathname'])[1:]
        print(read_path)
        phase = "train" if (count % 2) == 0 else "val"
        img = Image.open(read_path)
        objects = data['objects']
        for obj in objects:
            minr = obj['bounding_box']['minimum']['r']
            minc = obj['bounding_box']['minimum']['c']
            maxr = obj['bounding_box']['maximum']['r']
            maxc = obj['bounding_box']['maximum']['c']
            category = obj['category']
            img_crop = img.crop((minc,minr,maxc,maxr))
            if category == "red blood cell":
                name = "negative-" + phase + "-{:010d}".format(species[phase]["negative"])
                img_crop.save(os.path.join('../datasets',phase,"negative",name+".jpg"))
                print(name)
                species[phase]["negative"] += 1
            elif category == "ring" or category == "trophozoite" or category == "schizont" or category == "gametocyte":
                name = specie + "-" + phase + "-{:010d}".format(species[phase][specie])
                img_crop.save(os.path.join('../datasets',phase,specie,name+".jpg"))
                print(name)
                species[phase][specie] += 1
                if phase == "train":
                    img_crop = np.array(img_crop)
                    for deg in degs:
                        image_aug = iaa.Affine(rotate=deg,mode="reflect")(image=img_crop)
                        image_aug = Image.fromarray(image_aug)
                        name = specie + "-" + phase + "-{:010d}".format(species[phase][specie])
                        image_aug.save(os.path.join('../datasets',phase,specie,name+".jpg"))
                        print(name)
                        species[phase][specie] += 1
        count += 1
```
